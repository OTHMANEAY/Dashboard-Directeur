<T1> Analyse de Qualit√© des Donn√©es</T1>

```python
# valeurs manquantes par colonne
valeurs_nulles = df.isnull().sum()
taux_valeurs_nulles = round((df.isnull().sum() / len(df)) * 100, 2)

df_nulles = pd.DataFrame({
    'Colonne': valeurs_nulles.index,
    'Nb_Manquants': valeurs_nulles.values,
    'Pourcentage': taux_valeurs_nulles.values
}).sort_values('Nb_Manquants', ascending=False)

print(df_nulles[df_nulles['Nb_Manquants'] > 0])
```

```python
# Visualisation des valeurs manquantes
plt.figure(figsize=(12, 6))
df_nulles_plot = df_nulles[df_nulles['Nb_Manquants'] > 0].sort_values('Pourcentage')
plt.barh(df_nulles_plot['Colonne'], df_nulles_plot['Pourcentage'], color='coral')
plt.xlabel('Pourcentage de valeurs manquantes (%)')
plt.title('Distribution des Valeurs Manquantes par Colonne')
plt.axvline(x=5, color='red', linestyle='--', label='Seuil critique 5%')
plt.legend()
plt.tight_layout()
plt.show()
```

<T1> Analyse des Incoh√©rences</T1>

```python
# Valeur negatives pour Price Per Unit
negative_price = df[df['Price Per Unit'] <- 0]
print(f"Nombre de lignes : {len(negative_price)} ({len(negative_price)/len(df)*100:.2f}%)")
print(negative_price[['Transaction ID', 'Item', 'Price Per Unit']].head())
```

```python
# Valeurs n√©gatives pour Quantity
negative_qty = df[df['Quantity'] <= 0]
print(f"Nombre de lignes : {len(negative_qty)} ({len(negative_qty)/len(df)*100:.2f}%)")
print(negative_qty[['Transaction ID', 'Item', 'Quantity', 'Total Spent']].head())
```

```python
# Valeurs n√©gatives pour Total Spent
negative_total_spent = df[df['Total Spent'] <= 0]
print(f"Nombre de lignes : {len(negative_total_spent)} ({len(negative_total_spent)/len(df)*100:.2f}%)")
print(negative_total_spent[['Transaction ID', 'Item', 'Quantity', 'Total Spent']].head())
```

```python
# Verifier Total Spent est ce qu'il est rempli par des valeurs correctes (Quantity * Price Per Unit)

#1- Filtrer les lignes o√π Price Per Unit et Quantity ne sont pas null
df_valid = df[df['Price Per Unit'].notna() & df['Quantity'].notna()]

#2- Recalculer Total Spent pour ces lignes
df_valid['Total Spent Calculated'] = df_valid['Price Per Unit'] * df_valid['Quantity']

#3- Trouver les indices o√π le Total Spent original est diff√©rent du recalcul√©
diff_indices = df_valid[df_valid['Total Spent'] != df_valid['Total Spent Calculated']].index

# Afficher les indices
print(diff_indices)
```

```python
# Conversion de la colonne Date
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
```

```python
# V√©rifier les dates invalides
invalid_dates = df['Transaction Date'].isnull().sum()
print(f"Dates invalides : {invalid_dates}")
```

```python
# P√©riode couverte( Dates futures)
print(f"\nP√©riode couverte :")
print(f"-Date min : {df['Transaction Date'].min()}")
print(f"-Date max : {df['Transaction Date'].max()}")
print(f"-Dur√©e : {(df['Transaction Date'].max() - df['Transaction Date'].min()).days} jours")
```

```python
# D√©tection de trous temporels
date_range = pd.date_range(start=df['Transaction Date'].min(), end=df['Transaction Date'].max(), freq='D')
missing_dates = date_range.difference(df['Transaction Date'].dropna())
print(f"\nJours sans aucune vente : {len(missing_dates)}")
if len(missing_dates) > 0:
    print("Exemples :", missing_dates[:10].tolist())
```

```python
# Distribution des ventes par mois
monthly_sales = df.groupby(df['Transaction Date'].dt.to_period('M')).size()

plt.figure(figsize=(14, 6))
monthly_sales.plot(kind='bar', color='steelblue')
plt.title('Distribution des Transactions par Mois')
plt.xlabel('Mois')
plt.ylabel('Nombre de Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Distribution des ventes par annee
annual_sales = df.groupby(df['Transaction Date'].dt.year).size()

plt.figure(figsize=(14, 6))
annual_sales.plot(kind='bar', color='seagreen')
plt.title('Distribution des Transactions par Ann√©e')
plt.xlabel('Ann√©e')
plt.ylabel('Nombre de Transactions')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
